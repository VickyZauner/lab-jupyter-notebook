
<font colour='green'>**Indonesia**</font> is a country in Southeast Asia south off of Malaysia and Vietnam off the mainland of Asia between the Indian and the Pacific Ocean. The largest islands are *Sumatra, Java, Kalimantan, Sulawesi and the Indonesian part of New Guinea known as Papua New Guinea*, the propably most well known Island is the Island of *Bali*. One needs to keep in mind that the Indonesian Archepelago consits of **more than 17,508** officially listed islands!

The archepelago lies on the Pacific *"Ring of Fire"*, which means that a string of volcanoes runs through *Sumatra, Java, Bali and Nusa Tenggara* and then to the Banda Islands of Maluku to northeastern Sulawesi. Although frequent eruptions shake up and threaten the lives of the native population regularly, the ashes stemming from the outbreaks provide fertile grounds for the large variety of <font colour='green'>precious tropical fauna</font>. Also, climing vulcanoes at night and watching the sunset at **3000m** with instant coffee and fried bananas in hand is an exceptional experience.

Not only can you find beautiful tropical plants, flowers and fruits but Indonesia is also home to the whole wide range of tropical animals you suppose to be living in a jungle: Orang Utans, monkeys, crocodiles, tigers, elefants, birds, varans, ... and the list goes on.

The volcanic activity of the Islands also explains why some beaches are not as white or even pink as seen in many pictures of tropical coastlines but why certain beaches have black sand. Regardless of the color of sand, Indonesia is very well known for being home to **some of the best waves for surfing in the whole wide world**. And the wildlife underneath the surfave might be even more mindblowingly beautiful that the one found in the jungles.
